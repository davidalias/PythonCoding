```python
import re
```

```python
text = "Here is elon musk's number 9996667777, (123)-333-6776. If you have any questions on spacex, you can call him"
```

```python
pattern = '\(\d{3}\)-\d{3}-\d{4}|\d{10}'
```

```python
matches = re.findall(pattern, text)
matches
```

```python
text1 = "kadh;knfd-ldihdd-f;"
```

```python
pattern2 = '[^;-]+'
matches2 = re.findall(pattern2, text1)
matches2
```

##Exact Financial periods from a company's financial reporting

```python
text = '''
The gross cost of operating lease vehicles in FY2021 Q1 was $4.85 billion.
Tesla's employee count is 5000.
In previous quarter i.e., fY2020 Q4, it was $3 billion. FY2020 Q5
'''
```

```python
pattern = 'FY\d{4}.Q[1-4]'
matches = re.findall(pattern, text, flags=re.IGNORECASE)
matches
```

```python
pattern1 = 'FY(\d{4}.Q[1-4])'
matches1 = re.findall(pattern1, text, flags=re.IGNORECASE)
matches1
```

```python
pattern2 = '\$[0-9\.]+'
matches2 = re.findall(pattern2, text)
matches2
```

```python
text3 = """ 
hello@looseleafteas.com
WHOLESALE@LOOSELEAFTEAS.COM
summer@looseleafteas.com
tea_club_2026@looseleafteas.com
shop@loose-leaf.co.uk
"""
```

```python
pattern3 = '[a-zA-Z0-9_]+@'
matches = re.findall(pattern3, text3)
matches
```

```python
pattern4 = '[a-zA-Z0-9_]+@[a-zA-Z\-]+'
matches = re.findall(pattern4, text3)
matches
```

```python
pattern5 = '[a-zA-Z0-9_]+@[a-zA-Z-]+\.[a-zA-Z\.]+'
matches = re.findall(pattern5, text3)
matches
```
