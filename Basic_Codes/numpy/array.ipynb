```python
import numpy as np
```

```python
np1 = np.array([0,1,2,3,4,5])
print(np1)
```

```python
np1.shape
```

```python
#Range
np2 = np.arange(10)
np2
```

```python
#Step
np3 = np.arange(0,10,2)
print(np3)
```

```python
#Zeros
np4 = np.zeros(10)
print(np4)
```

```python
#multidimensional zeros
np5 = np.zeros((2,10))
print(np5)
```

```python
#Full
np6 = np.full((10), 6)
print(np6)
```

```python
#Multidimensional full
np7 = np.full((2,6), 5)
print(np7)
```

```python
#convert python list to numpy
mylist = [1,2,3,4,5]
np8 = np.array(mylist)
print(np8)
```

```python
print(np8[1])
```

```python
#slicing
np9 = np.arange(10)
print(np9)
print(np9[-3:-1])
print(np9[::2])
```

```python
#slicing 2D
np10 = np.array([
    [1,2,3,4], 
    [5,6,7,8]
    ])
print(np10)
print(np10[1,2])
print(np10[0:1,1:3])
print(np10[0:2, 1:3])
print(np10[:, 1:3])
```
