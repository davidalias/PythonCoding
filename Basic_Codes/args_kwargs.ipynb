```python
def average_cal(*scores):
    if len(scores) == 0:
        return 0
    average = sum(scores) / len(scores)
    return average
```

```python
print(average_cal(3,56,34))
print(average_cal(89,90,34,56,7,8,2,6))
print(average_cal())
```

```python
def describe_pet(breed, pet, **pet_info):
    print(f"{breed} is a {pet}") 
    for key, value in pet_info.items():
        print(f"{key} is {value}")
```

```python
describe_pet("whiskers", "cat", color = 'grey', age = 5)
```

```python
describe_pet("whiskers", "cat")
```

```python
def make_sandwich(*ingredients, **preferences):
    print(f"Sandwich is made of:")
    for item in ingredients:
        print(item)

    if preferences:
        print("Additional preferences:")

    for key, value in preferences.items():
        print(f"{key} is {value}")
```

```python
make_sandwich("turkey", "lettuce", "tomato", sauce = "mayo", toasted = True)
```

```python
make_sandwich("chicken", "cucumber", "cheese")
```
