Exception -> An event that interrupts the normal flow of a program. 
(ZeroDivisionError, TypeError, ValueError)
1. try 2. except 3. finally

"finally" block always execute, irrespective of any exception or not

```python
try:
    num = int(input("Enter number: "))
    print(1/num)
except ZeroDivisionError:
    print("You can't divide by zero")
except ValueError:
    print("Enter Only numbers please")
except Exception:
    print("Something is wrong!")
finally: #always execute, with or without exception
    print("Do some clean-up")

```

Exceptions are errors that occur while executing a program

Accident = Exception

Detour = Handling exception

```python
x = input("Enter number 1: ")
y = input("Enter number 2: ")
try:
    z = int(x)/int(y)
except Exception as e: #handling exception (accident) using detour
    print("Exception occured ", e)
    print("Exception type ", type(e).__name__)
    z = None
print("Division is", z)
```
