```python
import time
```

```python
def caketime():
    start = time.time()
    time.sleep(1)
    end = time.time()
    print("Time took for cake is", str((end-start)), "seconds")
```

```python
caketime()
```

```python
def matchacake():
    start = time.time()
    time.sleep(2)
    end = time.time()
    print("Time took for matcha cake is", str((end-start)), "seconds")
```

```python
matchacake()
```

```python
def time_calc(base_func):
    def enhanced_func(*args, **kwargs):
        start_time = time.time()
        result = base_func(*args, **kwargs)
        end_time = time.time()
        print("Time took for cake prep is", str((end_time-start_time)), "seconds")
        return result
    return enhanced_func

"""
You return enhanced_func (the function object), not enhanced_func() (the result of calling it), 
because a decorator's job is to produce a new callable that will be invoked laterâ€”
when the decorated function is actually called.

"""
```

```python
@time_calc
def caketime(sleep_time):
    print("Preparing cake...")
    time.sleep(sleep_time)
    return "good cake"
```

```python
caketime(2)
```

```python
@time_calc
def matchacake():
    print("Preparing cake...")
    time.sleep(1)
    return "good matcha cake"
```

```python
matchacake()
```

```python
caketime(sleep_time = 1)
```
