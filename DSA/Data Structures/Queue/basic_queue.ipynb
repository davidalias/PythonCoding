```python
class Queue:
    def __init__(self):
        self.q = []

    def insert(self, data):
        self.q.insert(0, data)

    def fifo(self):
        if len(self.q) == 0:
            return "No elements in the queue"
        self.q.pop()

    def display_q(self):
        return self.q
```

```python
q1 = Queue()
q1.insert({"a": 5, "b": 6})
q1.insert(5)
q1.insert(8)
q1.display_q()
```

```python
q1.fifo()
q1.display_q()
```
