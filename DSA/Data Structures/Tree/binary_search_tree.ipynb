```python
class BinarySearchTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def add_child(self, data):
        if self.data == data:
            return

        if data < self.data:
            #check for left node existence
            if self.left:
                self.left.add_child(data)
            else:
                self.left = BinarySearchTreeNode(data)

        else:
            if self.right:
                self.right.add_child(data)

            else:
                self.right = BinarySearchTreeNode(data)

    def in_place_traversal(self):
        values = []
        #visit left
        if self.left:
            values += self.left.in_place_traversal()
        
        #visit base
        values.append(self.data)

        #visit right
        if self.right:
            values += self.right.in_place_traversal()

        return values
    
    def search(self, val):
        if self.data == val:
            return True
        
        if val < self.data:
            if self.left:
                return self.left.search(val)
            else:
                return False
            
        if val > self.data:
            if self.right:
                return self.right.search(val)
            else:
                return False
            
    def find_max(self):
        if self.right == None:
            return self.data
        else:
            return self.right.find_max()
        
    def find_min(self):
        if self.left == None:
            return self.data
        else:
            return self.left.find_min()
        
    def find_number_of_childs(self, val):
        if self.data == val:
            if self.right and self.left:
                return 2
            elif self.left or self.right:
                return 1
            else:
                return 0
            
        elif val < self.data:
            if self.left:
                return self.left.find_number_of_childs(val)
            else:
                return None
            
        elif val > self.data:
            if self.right:
                return self.right.find_number_of_childs(val)

            else:
                return None
            
    def delete(self, val):

        if val < self.data:
            if self.left:
                self.left = self.left.delete(val)

        elif val > self.data:
            if self.right:
                self.right = self.right.delete(val)

        else:
            if self.left == None and self.right == None:
                return None
            if self.left is None:
                return self.right
            if self.right is None:
                return self.left
            
            min_val = self.right.find_min()
            self.data = min_val
            self.right = self.right.delete(min_val)

        return self       
            
```

```python
numbers = [5,10,25,30,2,7,8,3,1]
root= BinarySearchTreeNode(numbers[0])
for i in range(1, len(numbers)):
    root.add_child(numbers[i])
```
