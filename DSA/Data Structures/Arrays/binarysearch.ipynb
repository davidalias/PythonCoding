```python
def binary_search(sorted_arr, value):
    left, right = 0, len(sorted_arr) - 1
    while left <= right:
        mid = (right + left) //2
        if value > sorted_arr[mid]:
            left = mid + 1
        elif value < sorted_arr[mid]:
            right = mid - 1
        elif value == sorted_arr[mid]:
            return mid
        
    return None
```

```python
sorted_array = [2,6,9,10,23,78,95,100,110,145]
target = int(input(f"Enter the number from the list {sorted_array}: "))
print(f"The index of {target} is: ", binary_search(sorted_array, target) )
```
