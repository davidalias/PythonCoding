Given an integer array, move all zeroes to the end

```python
arr = [0,1,0,3,12]
```

```python
left, right = 0, len(arr)-1
while left < right:
    if arr[left] == 0:
        if arr[right] != 0:
            arr[left], arr[right] = arr[right], arr[left]
            left += 1
            right -= 1
        else:
            right -= 1
    else:
        left += 1
```

Given an array of vertical lines, find two lines that can trap the most water

```python
arr1 = [1,8,6,2,5,4,8,3,7]
```

```python
def find_area(arr):
    result = 0
    left, right = 0, len(arr)-1
    while left < right:
        area = min(arr[left], arr[right]) * (right-left)
        result = max(result, area)
        if arr[left] == arr[right]:
            if arr[left-1] > arr[right+1]:
                left += 1
            else:
                right -= 1
        elif arr[left] < arr[right]:
            left += 1
        else:
            right -= 1
    return result
```

```python
print(f"The maximum area between {arr1} is: ", find_area(arr1))
```
