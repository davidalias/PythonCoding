#Arrays

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4])
matrix = np.array([[1,2],[3,4]])
zeros = np.zeros((2,3))
ones = np.ones((2,2))
range_arr = np.arange(0,10,2)
linspace_arr = np.linspace(0,1,5)
```

```python
print('Array', arr)
print('matrix', matrix)
print("zeros", zeros)
print('ones', ones)
print('array_range', range_arr)
print('linspace_arr', linspace_arr)
```

Attributes

```python
print(arr.shape)
print(arr.ndim)
print(arr.size)
print(arr.dtype)
```

Indexing and slicing

```python
print('first element',arr[0])
print('last element', arr[-1])
print('slice', arr[1:3])
print('row0, column1', matrix[0,1])
print('allrows, column1', matrix[:,1])
```

Operations

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print('sum of two arrays', a+b)
print('multiply two arrays', a*b)
```

Scalar

```python
arr*2
```

Broadcasting

```python
matrix + 10
```

Useful Functions

```python
print('sum of an array', np.sum(arr))
print("mean of an array", np.mean(arr))
print("max value of an array", np.max(arr))
print("min value of an array", np.min(arr))
```

Reshaping and Flattening

```python
arr = np.arange(1,7)
reshaped = arr.reshape(2,3)
flatten = reshaped.flatten()
print('array',arr)
print('reshape', reshaped)
print('flatten', flatten)
```

Stacking

```python
a = np.array([1,2])
b = np.array([3,4])
print('vstack',np.vstack((a,b)))
print('hstack',np.hstack((a,b)))
```

Boolean Indexing

```python
arr = np.array([10,20,30,40])
print(arr[arr>20])
```

Random Arrays

```python
uniform = np.random.rand(3,3)
normal_dist = np.random.randn(3,3)
integ = np.random.randint(1,10, size = (2,3))
print('uniform distri', uniform)
print('normal distri', normal_dist)
print('integer', integ)
```

Linear Algebra

```python
a = np.array([[1,2], [3,4]])
b = np.array([[5,6], [7,8]])
print('dot matrix', np.dot(a,b))
print('Inverse', np.linalg.inv(a))
print('Determinant', np.linalg.det(a))
```

Copy and View

```python
arr = np.array([1,2,3])
view = arr.view()
copy = arr.copy()
print("view", view)
print('copy', copy)
```
