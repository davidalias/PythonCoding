Maximum sum subarray of size 5 (Fixed window problem)

```python
a = [8,3,-2,4,5,-1,0,5,3,9,-6]
```

```python
l = 0
s = sum(a[:5])
best = s
for r in range(5,len(a)-1):
    s = s + a[r] - a[r-5]
    best = max(best, s)
```

```python
print("The subarray of length 5 with max sum is:", best)
```

Longest subarray with sum less than 15

```python
arr = [4,5,2,0,1,8,12,3,6,9]
```

```python
best, left, right = 0,0,0
add, best_sum = 0, 0
```

```python
while right < len(arr):   
    add += arr[right]
    right += 1
    while add >= 15:
        add = add - arr[left]
        left += 1
    best = max(best, right-left)
    best_sum = max(best_sum, add)
```

```python
print(f"The longest subaray is of length {best}, and the sum of this subarray is {best_sum}")
```
