```python
def quicksort(list_1):

    if len(list_1) <= 1:
        return list_1
    
    else:
        pivot = list_1.pop()
        values_lower = [x for x in list_1 if x < pivot]
        values_higher = [x for x in list_1 if x >= pivot]
        return quicksort(values_lower) + [pivot] + quicksort(values_higher)
```

```python
print(f"Quicksort of array {list1} is", quicksort(list1))
```

In-place Method

```python
def partition(arr, low , high):
    pivot = arr[high]
    i = low -1
    for j in range(low, high):
        if arr[j] <= pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
        

    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i+1
```

```python
def quicksort(arr, low, high):
    if low < high:
        position = partition(arr, low, high)
        quicksort(arr, low, position - 1)
        quicksort(arr, position + 1, high)
```

```python
arr = [2,5,3,8,3,4,9,1,7]
print("Unsorted Array is:", arr)
quicksort(arr, 0, len(arr)-1)
print("Sorted array is:", arr)
```
