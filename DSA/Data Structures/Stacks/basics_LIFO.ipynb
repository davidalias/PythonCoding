```python
class Stack:
    def __init__(self):
        self.values = []

    def push(self, value):
        self.values = self.values + [value]

    def pop(self):
        self.values.pop()

    def peek(self):
        return self.values[-1]
    
    def display_stack(self):
        return self.values
```

```python
s1 =Stack()
s1.push(3)
s1.push(4)
s1.push(5)
s1.push(6)
print(s1.display_stack())
print(s1.peek())
```

```python
s1.pop()
s1.display_stack()
```
