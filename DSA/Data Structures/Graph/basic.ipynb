```python
class Graph:
    def __init__(self, edges):
        self.edges = edges
        self.graph_dict = {}
        for start, dest in self.edges:
            if start not in self.graph_dict:
                self.graph_dict[start] = [dest]
            else:
                self.graph_dict[start].append(dest)

    def printdict(self):
        print(self.graph_dict)
```

```python
routes = [
    ("Mumbai", "Paris"),
    ("Mumbai", "Dubai"),
    ("Paris", "Dubai"),
    ("Paris", "New York"),
    ("Dubai", "New York"),
    ("New York", "Toronto")
]
```

```python
g1 = Graph(routes)
g1.printdict()
```
