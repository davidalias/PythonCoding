```python
class Graph:
    def __init__(self, edges):
        self.edges = edges
        self.graph_dict = {}
        for start, dest in self.edges:
            if start not in self.graph_dict:
                self.graph_dict[start] = [dest]
            else:
                self.graph_dict[start].append(dest)

    def print_edges(self):
        print(self.graph_dict)

    def get_paths(self, start, end, path = []):
        path = path + [start]
        
        if start == end:
            return [path]
        
        if start not in self.graph_dict:
            return None
        
        all_paths = []
        for node in self.graph_dict[start]:
            if node not in path:
                new = self.get_paths(node, end, path)
                all_paths.extend(new)

        return all_paths
                          
```

```python
routes = [
    ("Mumbai", "Paris"),
    ("Mumbai", "Dubai"),
    ("Paris", "Dubai"),
    ("Paris", "New York"),
    ("Dubai", "New York"),
    ("New York", "Toronto")
]
```

```python
g1 = Graph(routes)
g1.print_edges()
```
