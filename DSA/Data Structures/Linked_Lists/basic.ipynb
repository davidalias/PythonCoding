```python
class Node:
    def __init__(self,x):
        self.value =x
        self.next = None
```

```python
p1 =Node(10)
p2 = Node(20)
p3 = Node(30)
```

```python
p1.next = p2
p2.next  = p3
```

```python
flow = p1
while flow != None:
    print(flow.value)
    flow = flow.next #go to the next node
```
