```python
class Node:
    def __init__(self, data = None, next = None):
        self.data = data
        self.next = next
```

```python
node1 = Node(1)
node2 = Node(2)
node3 = Node(3)
node4 = Node(4)
```

```python
node1.next = node2
node2.next = node3
node3.next = node4
node4.next = node2
```

```python
def ll_is_cycle(node1 = node1):
    slow, fast = node1, node1
    while slow and fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
        
    return False
```

```python
ll_is_cycle()
```
